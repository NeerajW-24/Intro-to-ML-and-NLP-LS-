```python
import numpy as np
```

```python
arr = np.random.randint(1, 51, (5, 4))
print(arr)
```

```python
anti_diagonal = [arr[i][3 - i] for i in range(min(arr.shape))]
print(anti_diagonal)
```

```python
row_max = np.max(arr, axis=1)
print(row_max)
```

```python
mean_val = np.mean(arr)
new_arr = arr[arr <= mean_val]
print(new_arr)
```

```python
def numpy_boundary_traversal(matrix):
    top = matrix[0]
    right = matrix[1:-1, -1]
    bottom = matrix[-1][::-1]
    left = matrix[-2:0:-1, 0]
    return list(top) + list(right) + list(bottom) + list(left)
```

```python
print(numpy_boundary_traversal(arr))
```
