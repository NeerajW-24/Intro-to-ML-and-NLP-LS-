```python
import numpy as np
```

```python
arr = np.random.uniform(0, 10, 20)
arr = np.round(arr, 2)
print(arr)
```

```python
print(np.min(arr))
print(np.max(arr))
print(np.median(arr))
```

```python
arr[arr < 5] = arr[arr < 5] ** 2
print(arr)
```

```python
def numpy_alternate_sort(array):
    sorted_arr = np.sort(array)
    result = []
    i, j = 0, len(sorted_arr) - 1
    while i <= j:
        result.append(sorted_arr[i])
        if i != j:
            result.append(sorted_arr[j])
        i += 1
        j -= 1
    return np.array(result)
```

```python
print(numpy_alternate_sort(arr))
```
