```python
import pandas as pd
import numpy as np
```

```python
names = ['Student'+str(i) for i in range(1, 11)]
subjects = ['Math', 'Science', 'English', 'History', 'Geography'] * 2
scores = np.random.randint(50, 101, 10)
```

```python
df = pd.DataFrame({
    'Name': names,
    'Subject': subjects,
    'Score': scores,
    'Grade': ''
})
```

```python
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
```

```python
df['Grade'] = df['Score'].apply(assign_grade)
print(df.sort_values(by='Score', ascending=False))
```

```python
print(df.groupby('Subject')['Score'].mean())
```

```python
def pandas_filter_pass(dataframe):
    return dataframe[dataframe['Grade'].isin(['A', 'B'])]
```

```python
print(pandas_filter_pass(df))
```
